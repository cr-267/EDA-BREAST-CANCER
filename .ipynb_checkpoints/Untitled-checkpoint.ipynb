```python
x_train,x_test,y_train,y_test = train_test_split(df,target,test_size=0.2,random_state=32)
x_train.shape
```

# Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()

x_train_ss = ss.fit_transform(x_train)
x_test_ss = ss.fit_transform(x_test)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
import tensorflow as tf
tf.random.set_seed(3)
```

```python
model_nn = tf.keras.Sequential([
                            tf.keras.layers.Flatten(input_shape = (30,)),
                            tf.keras.layers.Dense(15,activation='relu'),
                            tf.keras.layers.Dense(12,activation='relu'),
                            tf.keras.layers.Dense(8,activation='relu'),
                            tf.keras.layers.Dense(2,activation = "sigmoid")
])
```

```python
model_nn.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics = ["accuracy"]
             )
```

```python
history = model_nn.fit(x_train_ss,y_train,validation_split=0.1,epochs=10)
```

```python
history.history['accuracy']
```

```python
history.history['loss']
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("Neural Network accuracy")
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(["training data","validation_data"])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title("model loss")
plt.xlabel('epoch')
plt.ylabel('loss')

plt.legend(["training data","validation_data"])
```

```python
loss_nn,accuracy = model_nn.evaluate(x_test_ss,y_test)
```

```python

models = {
    'RandomForest': RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42),
    'AdaBoost': AdaBoostClassifier(n_estimators=30, learning_rate=0.01, random_state=42),
    'GradientBoosting': GradientBoostingClassifier(n_estimators=50, learning_rate=0.01, max_depth=3, random_state=42),
    'XGBoost': xgb.XGBClassifier(n_estimators=50, learning_rate=0.01, max_depth=3, random_state=42),
    'LogisticRegression': LogisticRegression(max_iter=500),
    'SVM': SVC(C=1.0, kernel='rbf', probability=True, random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'DecisionTree': DecisionTreeClassifier(max_depth=5, random_state=42)
}




accuracy_list = []
loss_list = []
precision_list = []
recall_list = []
f1_list = []
auc_list = []


for name, model in models.items():
    model.fit(x_train_ss, y_train)
    y_pred = model.predict(x_test_ss)
    y_pred_prob = model.predict_proba(x_test_ss)[:, 1]
    
    accuracy = accuracy_score(y_test, y_pred)
    loss = log_loss(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    auc_value = roc_auc_score(y_test, y_pred_prob)
    
    accuracy_list.append(accuracy)
    loss_list.append(loss)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)
    auc_list.append(auc_value)
    
    





models["Neural Network"] = model_nn
accuracy_list.append(accuracy_nn)
loss_list.append(loss_nn)
precision_list.append(precision_nn)
recall_list.append(recall_nn)
f1_list.append(f1_nn)
auc_list.append(auc_nn)
```

```python
plt.figure(figsize=(10, 6))
for name, model in models.items():
    if name !="Neural Network" :
        y_pred_prob = model.predict_proba(x_test_ss)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
    else:
        y_pred_prob = model.predict(x_test_ss)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
# Random Forest Feature Importance
rf_model = models['RandomForest']
importances = rf_model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Feature Importance - Random Forest")
plt.bar(range(x_train_ss.shape[1]), importances[indices], align="center")
plt.xticks(range(x_train_ss.shape[1]), indices)
plt.show()
```

```python
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt

def plot_learning_curve(model, title, X, y):
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5))
    
    train_scores_mean = np.mean(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)

    plt.plot(train_sizes, train_scores_mean, label="Training score", color="darkorange")
    plt.plot(train_sizes, test_scores_mean, label="Cross-validation score", color="navy")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.legend(loc="best")
    plt.grid()

def plot_learning_curves_for_all(models, X, y):
    plt.figure(figsize=(15, len(models) * 5))  
    num_models = len(models)
    
    for i, (name, model) in enumerate(models.items(), 1):
        if name !="Neural Network":
            plt.subplot(num_models, 1, i)  
            plot_learning_curve(model, f'Learning Curve - {name}', X, y)
            plt.title(f'Learning Curve - {name}')
            
    
    plt.tight_layout()  
    plt.show()


plot_learning_curves_for_all(models, x_train_ss, y_train)
```
