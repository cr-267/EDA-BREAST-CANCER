```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.shape
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
data.describe()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})
data.head()
```

```python
# Step 2: Visualizing Diagnosis Distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='diagnosis', palette='coolwarm')
plt.title('Diagnosis Distribution (0 = Benign, 1 = Malignant)')
plt.xlabel('Diagnosis (0 = Benign, 1 = Malignant)')
plt.ylabel('Count')
plt.show()
```

```python
dara = data.drop(["Unnamed: 32"],axis=1,inplace=True)
data.drop(["id"],axis=1,inplace=True)
data.head()
```

```python
plt.figure(figsize=(16, 12))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
features_to_plot = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
```

```python
plt.figure(figsize=(12, 8))
for feature in features_to_plot:
    plt.subplot(2, 3, features_to_plot.index(feature) + 1)
    sns.histplot(data=data, x=feature, hue='diagnosis', bins=30, kde=True, palette='coolwarm', alpha=0.6)
    plt.title(f'Distribution of {feature} by Diagnosis')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
for feature in features_to_plot:
    plt.subplot(2, 3, features_to_plot.index(feature) + 1)
    sns.boxplot(data=data, x='diagnosis', y=feature, palette='Set3')
    plt.title(f'Boxplot of {feature} by Diagnosis')
    plt.xlabel('Diagnosis (0 = Benign, 1 = Malignant)')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data[['radius_mean', 'texture_mean', 'area_mean', 'smoothness_mean', 'diagnosis']], 
             hue='diagnosis', palette='coolwarm')
plt.suptitle('Pairplot of Key Features Colored by Diagnosis', y=1.02)
plt.show()
```

```python
features_mean = [col for col in data.columns if '_mean' in col]
features_worst = [col for col in data.columns if '_worst' in col]
```

```python
mean_worst_comparison = data[['diagnosis'] + features_mean + features_worst]

print("\nMean vs Worst Feature Comparison:")
print(mean_worst_comparison.describe())
```

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(data=data, x='radius_mean', y='area_mean', hue='diagnosis', palette='coolwarm', alpha=0.6)
plt.title('Radius Mean vs Area Mean by Diagnosis')
plt.xlabel('Radius Mean')
plt.ylabel('Area Mean')
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette='coolwarm', alpha=0.6)
plt.title('PCA of Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Diagnosis', labels=['Benign (0)', 'Malignant (1)'])
plt.show()
```
